# Stationarity and Weak Dependence

Given a stochastic process **X<sub>t</sub>**.

### Definiton of Mean Stationarity

* E[X<sub>t</sub>] = $\mu$ $\neq$ f(t)

* Var[X<sub>t</sub>] = $\sigma$<sup>2</sup> $\neq$ f(t)

* Cov[X<sub>t</sub>,X<sub>t+h</sub>] = f(h) $\neq$ f(t)

### Weak Dependency

Cov[X<sub>t</sub>,X<sub>t+h</sub>] = 0 when h -> $\infty\$ <br>
(or when the time point is at an infinite lag from another, there covariance tends to 0)

# Random Walk

AR(1) process with **$\rho$=1**

X<sub>t</sub> = X<sub>t-1</sub> + $\xi\$, 

where $\xi\$ belongs to N(0,1)  

```python
import numpy as np
import pandas as pd

X = []
start = 0

for _ in range(100):
    new = start + np.random.normal(0,1)
    X.append(new)
    start = new
    
X = pd.DataFrame(X,columns=["random_walk"])
X.plot()
```

```python
#Mean of a Random Walk
X.mean()
```
